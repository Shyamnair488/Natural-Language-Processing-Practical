#**Edit distance Algorithms**

```python
import nltk
from nltk.metrics.distance import edit_distance
```

```python
nltk.download('words')
```

```python
from nltk.corpus import words
```

```python
cor_word=words.words() # dictionary
```

```python
incor_word=['wong','hapy','clou']
print(incor_word)
```

```python
for word in incor_word:
  temp=[(edit_distance(word,w),w) for w in cor_word if w[0]==word[0]]
  print(sorted(temp,key=lambda val:val[0])[0][1])
```
