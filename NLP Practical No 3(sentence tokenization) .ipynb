Tokenization:Tokenization is the process of tokenizing or splitting a string, text into a list of tokens.

Types:
      1. Sentence tokenization or segmentation
      2. word tokenization




```python
import nltk
```

```python
nltk.download('punkt')
```

```python
data="In the heart of the bustling city, where skyscrapers touch the clouds and neon lights illuminate the streets, a sense of constant motion and energy permeates the air. People from all walks of life hurry along sidewalks, their footsteps echoing amidst the cacophony of honking cars and distant chatter. Amidst this urban chaos, hidden gems of serenity can be found – quaint cafes with the aroma of freshly brewed coffee, and hidden parks where nature's whispers offer a momentary escape. Each corner holds a story, waiting to be discovered by those who venture beyond the surface of this vibrant tapestry of life."
```

```python
Sentence_tokenization=nltk.sent_tokenize(data)
Sentence_tokenization
```

```python
Word_tokenization=nltk.word_tokenize(data)
Word_tokenization
```
