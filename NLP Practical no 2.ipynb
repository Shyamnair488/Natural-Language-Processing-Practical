**Stemming: The process of reducing a word to its word stem that affixes to suffixes and prefixes or the roots.**

**Types of stemming:
                  1. porter stemmer
                  2. snowball
                  3. lancaster
                  4. regex**

```python
import nltk
```

```python
nltk.download('punkt')
```

--------------------------------------------------------------------------------------------------------------------------------------------------------

**Porter  Stemmer**

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
data0="The quick brown Foxes are jumping over the lazy dog"
data0
```

```python
token_data=nltk.word_tokenize(data0)
print(token_data)
```

```python
Stem_data0=[ps.stem(word)for word in token_data]
```

```python
Stem_data_sentence0=" ".join(Stem_data0)
Stem_data_sentence0
```

```python
print("Before Sentence: ",data0)
print("After Sentence :",Stem_data_sentence0)
```

**SnowBall Stemmer**

```python
from nltk.stem import SnowballStemmer
```

```python
Sb=SnowballStemmer(language='english')
```
