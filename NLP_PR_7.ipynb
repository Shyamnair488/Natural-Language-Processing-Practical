**GRAMMAR CHECKING**

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
def check_grammar(sentence):
  words = nltk.word_tokenize(sentence)
  tagged_words = nltk.pos_tag(words)
  errors = []
  for word, pos_tag in tagged_words:
    if pos_tag not in ["VB", "VBD", "VBG", "VBN", "VBP", "VBZ", "JJ", "JJR", "JJS", "RB", "RBR", "RBS", "DT", "IN"]:
      errors.append(f"Grammar error: '{word}'")
  return errors
```

```python
if __name__ == "__main__":
  sentences=["I ARE USE NLP.",
            "THEY IS READ COMIC.",
            "I AM READING A BOOK.",
            "STUDENTS ARE LEARNING"]
```

```python
for sentence in sentences:
  errors = check_grammar(sentence)
  if errors:
    print(f"Sentence: '{sentence}'")
    for error in errors:
      print(f" - {error}")
```
