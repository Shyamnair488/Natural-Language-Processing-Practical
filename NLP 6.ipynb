**Aim : Write a python program to implement pos tagging**

```python
import nltk
```

```python
nltk.download('all')
nltk.download('punkt')
```

```python
Sent='''Good morning, sir, I hope you are feeling well today;
 as your dedicated and reliable assistant,
  I am here to provide you with my full support,
 whether it's answering your questions,
assisting with tasks, or simply making your day a bit easier and more enjoyable."
'''
Sent
```

```python
token_data=nltk.word_tokenize(Sent)
print(token_data)
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
postag_data=nltk.pos_tag(token_data)
tags=postag_data
print(tags)
```

```python
# Extract unique POS tags
unique_tags = set(tag for word, tag in tags)

# Count the number of unique POS tags
num_unique_tags = len(unique_tags)
```

```python
# Print the unique tags and the count
print("Unique POS Tags:", unique_tags)
print("Number of Unique POS Tags:", num_unique_tags)
```
