Language1 = English

```python
P1="Amidst the towering shelves of an antiquarian bookstore, the scent of aged parchment and leather bindings enveloped me like a comforting embrace. Dust motes danced lazily in the slivers of sunlight that filtered through the small, cracked windows. Rows upon rows of weathered books, their spines adorned with fading titles and ornate designs, stood as silent sentinels to the wisdom contained within their pages. Each volume seemed to hold the whispers of generations past, waiting to share their tales and knowledge with the curious reader who dared to open their covers. In this haven of forgotten stories and forgotten times, I found myself lost in a world where the past and present coalesced, and the magic of literature transcended the boundaries of time and space."
P1
```

```python
Sentence_tokenization=sent_tokenize(P1,language='english')
Sentence_tokenization
```

```python
Word_tokenization=word_tokenize(P1,language='english')
print(Word_tokenization)
```
